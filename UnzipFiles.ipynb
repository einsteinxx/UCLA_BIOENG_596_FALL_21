```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import gc  #debug memory leaks in matplotlib
import csv #read in description files
import random #used to select random slice for patches
import cv2

################################################################################
#ULTRASOUND NEEDS
import PIL

from matplotlib import image
from ast import literal_eval #used to break out bounding boxes from strings
import time

from zipfile import ZipFile

################################################################################

eps = 1e-16 #keep div by 0 away



#
# Read Data from google drive
#
from google.colab import drive #for loading gdrive data
from google.colab import files



################################################################################
# Load data from google drive
################################################################################
drive.mount('/content/gdrive')

#
# Setup of input and output drives
#


#storage areas 
zip_dir = '/content/gdrive/Shareddrives/BreastUS/Annotated data/Batch2'

output_dir = '/content/gdrive/Shareddrives/BreastUS/Annotated data/unzipped_Batch2'
local_dir = '/content/gdrive/My Drive/BreastUS' #for local storage
```

```python
print(zip_dir)
zip_files = os.listdir(zip_dir)


#
# Make top level output file for Batch2 subdirectories
#
if(os.path.isdir(output_dir)== False):
    os.mkdir(output_dir)
    print('made output dir')
else:
    print('output directory already exists')


#
# Unzip each main file and place the contents in the designated folder
# -- this reuses the long .zip name as the new output folder, to be 
#    consistent with previous setups

for zip_name in zip_files:
    #print(zip_name)
    new_dir = os.path.join(output_dir,zip_name)
    full_zip_name = os.path.join(zip_dir,zip_name)
    if (os.path.isdir(new_dir) == False):
        os.mkdir(new_dir)
    else:
        print('*** dir already created: ',new_dir)
        #get a small list of the files in there
        folder_contents = os.listdir(new_dir)
        if (len(folder_contents) > 0):
            #found something already there, pass this one 
            continue

    
    with ZipFile(full_zip_name, 'r') as zipObj:
        # Extract all the contents of zip file in different directory
        zipObj.extractall(new_dir) #os.path.join(new_dir,'default'))
    print('-------------done extracting ', new_dir)
    files_found = os.listdir(new_dir)
    if (len(files_found) ==0):
        print('!!!!!!! NO OUTPUT FILES FOUND FOR ', zip_name)
    else:
        print('# files found in unzipped area: ', os.listdir(new_dir))
    
```
